```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# CSV file load pannurom
df = pd.read_csv(r'C:\Users\Admin\Downloads\rfm\data.csv', encoding='latin1')

print(df.shape)
print(df.head())
```

```python
# Missing values check pannurom
print(df.isnull().sum())
print("\nData Shape:", df.shape)
```

```python
# Cleaning
df = df.dropna(subset=['Customer ID'])  # CustomerID illatha rows delete
df = df[df['Quantity'] > 0]             # Returns remove pannurom
df = df[df['Price'] > 0]               # Free items remove pannurom

# Revenue column create pannurom
df['Revenue'] = df['Quantity'] * df['Price']

print("Clean data rows:", len(df))
print(df.head())
```

```python
# Cleaning
df = df.dropna(subset=['CustomerID'])   # CustomerID illatha rows delete
df = df[df['Quantity'] > 0]             # Returns remove pannurom
df = df[df['UnitPrice'] > 0]           # Free items remove pannurom

# Revenue column create pannurom
df['Revenue'] = df['Quantity'] * df['UnitPrice']

print("Clean data rows:", len(df))
print(df.head())
```

```python
from datetime import datetime

# Reference date set pannurom
ref_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# RFM Calculate pannurom
rfm = df.groupby('CustomerID').agg(
    Recency   = ('InvoiceDate', lambda x: (ref_date - x.max()).days),
    Frequency = ('InvoiceNo', 'nunique'),
    Monetary  = ('Revenue', 'sum')
).reset_index()

print(rfm.shape)
print(rfm.head(10))
```

```python
# InvoiceDate datetime format-a convert pannurom
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Reference date set pannurom
ref_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# RFM Calculate pannurom
rfm = df.groupby('CustomerID').agg(
    Recency   = ('InvoiceDate', lambda x: (ref_date - x.max()).days),
    Frequency = ('InvoiceNo', 'nunique'),
    Monetary  = ('Revenue', 'sum')
).reset_index()

print(rfm.shape)
print(rfm.head(10))
```

```python
# 1-5 Score assign pannurom
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

# RFM Score combine pannurom
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# Segment assign pannurom
def assign_segment(row):
    r = int(row['R_Score'])
    f = int(row['F_Score'])
    m = int(row['M_Score'])
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champion'
    elif r >= 3 and f >= 3:
        return 'Loyal Customer'
    elif r >= 4 and f <= 2:
        return 'New Customer'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2:
        return 'Hibernating'
    else:
        return 'Potential Loyalist'

rfm['Segment'] = rfm.apply(assign_segment, axis=1)

print(rfm['Segment'].value_counts())
print(rfm.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Segment Distribution â Pie Chart
plt.figure(figsize=(8,6))
rfm['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Customer Segment Distribution')
plt.ylabel('')
plt.show()

# 2. Revenue by Segment â Bar Chart
plt.figure(figsize=(10,6))
seg_revenue = rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
sns.barplot(x=seg_revenue.index, y=seg_revenue.values, palette='Blues_d')
plt.title('Revenue by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
import sqlite3

# Database create pannurom
conn = sqlite3.connect(r'C:\Users\Admin\Downloads\rfm\rfm_database.db')

# RFM data SQL-la load pannurom
rfm.to_sql('rfm_table', conn, if_exists='replace', index=False)

# SQL Query run pannurom
query = """
SELECT Segment,
       COUNT(*) AS Customer_Count,
       ROUND(AVG(Monetary), 2) AS Avg_Spend,
       ROUND(SUM(Monetary), 2) AS Total_Revenue
FROM rfm_table
GROUP BY Segment
ORDER BY Total_Revenue DESC
"""

result = pd.read_sql_query(query, conn)
print(result)
```
